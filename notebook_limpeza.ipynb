```python
df = df.dropna()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
top_10 = df.nlargest(10, "Desmatado 2022")

print(top_10[["Municipio", "Estado", "Desmatado 2022"]])


```
